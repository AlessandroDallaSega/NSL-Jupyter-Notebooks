```python
#per termalizzare un sistema solido servono circa 300 step partendo da una temperatura iniziale di 1.54
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

n, a = np.loadtxt("temperature_termal.dat", usecols=(0,1), unpack='true')
plt.plot(n, a)
plt.xlim(0,2000)
plt.show()
```

```python

#grafici solido

n, E, err_E = np.loadtxt("total_energy_solid.dat", usecols=(0,2,3), unpack='true')
n, K, err_K = np.loadtxt("kinetic_energy_solid.dat", usecols=(0,2,3), unpack='true')
n, P, err_P = np.loadtxt("potential_energy_solid.dat", usecols=(0,2,3), unpack='true')
n, T, err_T= np.loadtxt("temperature_solid.dat", usecols=(0,2,3), unpack = 'true') 
n, p, err_p = np.loadtxt("pressure_solid.dat", usecols=(0,2,3), unpack='true')

figure, axis = plt.subplots(2, 3)
figure.set_figwidth (15)
figure.set_figheight (10)

axis[1,2].axis('off')

axis [0,0].errorbar( n, E, yerr = err_E)
axis [0,0].set_title("tot energy")

axis [0,1].errorbar( n, K, yerr = err_K)
axis [0,1].set_title("kinetic e")

axis [0,2].errorbar(n, P, yerr = err_P)
axis [0,2].set_title("potential e")

axis [1,0].set_title("temperature")
axis [1,0].errorbar(n, T, yerr = err_T)

axis [1,1].set_title("pressure")
axis [1,1].errorbar(n, p, yerr = err_p)

plt.show()

```
